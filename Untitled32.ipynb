```python
%%writefile security.py
import hashlib
import re
import secrets

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Must be at least 8 characters"
    if not re.search(r"[A-Z]", password):
        return "Weak: Add uppercase letter"
    if not re.search(r"[0-9]", password):
        return "Weak: Add a number"
    if not re.search(r"[!@#$%^&*]", password):
        return "Weak: Add special character"
    return "Strong"

def generate_token():
    return secrets.token_hex(16)

def validate_username(username):
    return re.match("^[a-zA-Z0-9_]+$", username)
```

```python
%%writefile main.py
import json
import os
from security import *

DB_FILE = "database.json"
LOG_FILE = "logs.txt"

def load_data():
    if not os.path.exists(DB_FILE):
        return {}
    with open(DB_FILE, "r") as f:
        return json.load(f)

def save_data(data):
    with open(DB_FILE, "w") as f:
        json.dump(data, f, indent=4)

def log_event(event):
    with open(LOG_FILE, "a") as f:
        f.write(event + "\n")

def register():
    data = load_data()

    username = input("Enter username: ")
    if not validate_username(username):
        print("Invalid username!")
        return

    if username in data:
        print("User already exists!")
        return

    password = input("Enter password: ")
    strength = check_password_strength(password)
    print(strength)

    if "Weak" in strength:
        return

    data[username] = {
        "password": hash_password(password),
        "recovery_token": generate_token()
    }

    save_data(data)
    log_event(f"REGISTER: {username}")
    print("Registered successfully!")

def login():
    data = load_data()

    username = input("Username: ")
    password = input("Password: ")

    if username in data and data[username]["password"] == hash_password(password):
        print("Login successful!")
        log_event(f"LOGIN SUCCESS: {username}")
    else:
        print("Login failed!")
        log_event(f"LOGIN FAIL: {username}")

def main():
    while True:
        print("\n1.Register\n2.Login\n3.Exit")
        choice = input("Choose: ")

        if choice == "1":
            register()
        elif choice == "2":
            login()
        elif choice == "3":
            break

if __name__ == "__main__":
    main()
```
